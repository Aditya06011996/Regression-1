Q2. Discuss the assumptions of linear regression. How can you check whether these assumptions hold in
a given dataset?

Ans - Linear regression relies on several assumptions for its validity. It's important to check these assumptions to ensure the reliability of the regression model. Here are the key assumptions:

1. **Linearity:**
   - **Assumption:** The relationship between the independent variable(s) and the dependent variable is linear.
   - **Check:** Examine scatterplots of the variables or use residual plots to assess linearity visually.

2. **Independence of Residuals:**
   - **Assumption:** The residuals (the differences between observed and predicted values) are independent of each other.
   - **Check:** Plot residuals against predicted values or time (if applicable) to identify patterns or trends.

3. **Homoscedasticity (Constant Variance of Residuals):**
   - **Assumption:** The spread of residuals is constant across all levels of the independent variable(s).
   - **Check:** Scatterplots of residuals against predicted values can reveal any cone-shaped or fan-shaped patterns, which indicate heteroscedasticity.

4. **Normality of Residuals:**
   - **Assumption:** The residuals are normally distributed.
   - **Check:** Histograms or Q-Q plots of residuals can be examined. Additionally, statistical tests like the Shapiro-Wilk test or the Kolmogorov-Smirnov test can be used to test for normality.

5. **No Perfect Multicollinearity:**
   - **Assumption:** There is no exact linear relationship between the independent variables.
   - **Check:** Calculate variance inflation factor (VIF) for each independent variable. High VIF values (typically above 10) may indicate multicollinearity.

6. **No Autocorrelation of Residuals:**
   - **Assumption:** The residuals are not correlated with each other (no pattern in the residuals over time).
   - **Check:** Plot residuals against time (if applicable) or use statistical tests such as the Durbin-Watson test.

### How to Check Assumptions:

1. **Visual Inspection:**
   - Examine scatterplots, residual plots, and histograms.
   - Look for patterns, linearity, and normality.

2. **Statistical Tests:**
   - Use statistical tests for normality (e.g., Shapiro-Wilk).
   - Calculate VIF for multicollinearity.
   - Perform tests for autocorrelation (e.g., Durbin-Watson).

3. **Residual Analysis:**
   - Examine residuals for any patterns or trends.
   - Check for constant variance in residuals.

4. **Influence and Outlier Detection:**
   - Identify influential points or outliers that may disproportionately affect the regression model.

It's crucial to note that violations of these assumptions don't necessarily invalidate the model but may affect its interpretation and predictions. In some cases, transformations or alternative modeling approaches may be considered to address violations.

Q6. Explain the concept of multicollinearity in multiple linear regression. How can you detect and
address this issue?

Ans - **Multicollinearity in Multiple Linear Regression:**

Multicollinearity is a phenomenon in multiple linear regression where two or more independent variables in a model are highly correlated. This high correlation can cause problems in the regression analysis, making it challenging to isolate the individual effect of each independent variable on the dependent variable.

**Issues Caused by Multicollinearity:**

1. **Unstable Coefficients:** The coefficients become sensitive to small changes in the data, leading to instability in the estimation.

2. **Increased Standard Errors:** The standard errors of the coefficients can be inflated, making it difficult to assess the statistical significance of individual variables.

3. **Reduced Precision:** Multicollinearity reduces the precision of the estimated coefficients, making it harder to identify the true relationships between variables.

**Detecting Multicollinearity:**

Several methods can be used to detect multicollinearity:

1. **Correlation Matrix:**
   - Examine the correlation matrix between independent variables. High correlation coefficients (close to 1) indicate potential multicollinearity.

2. **Variance Inflation Factor (VIF):**
   - Calculate the VIF for each independent variable. VIF measures how much the variance of an estimated regression coefficient increases if your predictors are correlated. A VIF greater than 10 is often considered a sign of multicollinearity.

3. **Tolerance:**
   - Tolerance is another measure related to VIF. It is the reciprocal of VIF (\(1/\text{VIF}\)), and low tolerance values indicate high multicollinearity.

**Addressing Multicollinearity:**

1. **Remove Redundant Variables:**
   - If two variables are highly correlated, consider removing one of them from the model.

2. **Combine Variables:**
   - If possible, create new variables that combine the information from highly correlated variables.

3. **Increase Sample Size:**
   - Increasing the sample size can sometimes help mitigate multicollinearity.

4. **Regularization Techniques:**
   - Techniques like Ridge Regression or Lasso Regression introduce a penalty term to the coefficients, which can help stabilize them and reduce multicollinearity.

5. **Principal Component Analysis (PCA):**
   - PCA can transform the original variables into a set of uncorrelated variables (principal components) to address multicollinearity.

6. **VIF Threshold:**
   - Use a threshold for VIF (e.g., VIF < 10) and exclude variables that exceed this threshold.

**Caution:**
Addressing multicollinearity should be done carefully, and the chosen solution depends on the specific context and goals of the analysis. Removing variables without a strong theoretical basis or arbitrarily combining variables can lead to biased results. It's crucial to understand the underlying causes of multicollinearity and choose an appropriate strategy based on the characteristics of the data and the goals of the analysis.

Q8. What are the advantages and disadvantages of polynomial regression compared to linear
regression? In what situations would you prefer to use polynomial regression?

**Advantages of Polynomial Regression:**

1. **Capturing Non-Linear Patterns:**
   - Polynomial regression can capture non-linear relationships between variables, allowing for more flexible modeling of complex patterns in the data.

2. **Higher Order Relationships:**
   - It can represent relationships beyond a straight line by introducing polynomial terms of higher order.

3. **Increased Flexibility:**
   - Polynomial regression is more flexible than linear regression and can fit a wider range of data patterns.

**Disadvantages of Polynomial Regression:**

1. **Overfitting:**
   - Using high-degree polynomials can lead to overfitting, where the model captures noise in the data rather than the underlying pattern. This can result in poor generalization to new, unseen data.

2. **Increased Complexity:**
   - As the degree of the polynomial increases, the model becomes more complex, and the interpretation of coefficients becomes challenging.

3. **Loss of Interpretability:**
   - With higher-degree polynomials, it becomes more difficult to interpret the individual contributions of each term to the overall prediction.

4. **Sensitivity to Outliers:**
   - Polynomial regression can be sensitive to outliers, particularly with higher-degree polynomials, leading to inflated influence of outliers on the model.

5. **Computational Complexity:**
   - The computation of higher-degree polynomial models can be computationally expensive, especially for large datasets.

**Situations for Using Polynomial Regression:**

1. **Non-Linear Relationships:**
   - When there is evidence of a non-linear relationship between the independent and dependent variables, polynomial regression may be more appropriate.

2. **Complex Data Patterns:**
   - In situations where the relationship between variables is complex and cannot be adequately captured by a linear model, polynomial regression offers a more flexible alternative.

3. **Curvilinear Trends:**
   - When the data exhibits curvilinear trends, polynomial regression with an appropriate degree can provide a better fit.

4. **Limited Assumptions:**
   - Linear regression assumes a linear relationship, which may not hold in many real-world scenarios. In such cases, polynomial regression relaxes this assumption.

5. **Understanding Trade-offs:**
   - When the trade-off between model complexity and fit to the data is carefully considered. Using lower-degree polynomials may provide a good balance between capturing non-linearity and avoiding overfitting.

In summary, while polynomial regression offers increased flexibility for capturing non-linear relationships, it comes with the trade-off of increased complexity and potential overfitting. It is essential to carefully assess the data, consider the model's interpretability, and use techniques like regularization to mitigate overfitting in higher-degree polynomial models. Polynomial regression is most suitable when it aligns with the underlying patterns in the data and when linear regression is not sufficient to capture the complexity of the relationship.
